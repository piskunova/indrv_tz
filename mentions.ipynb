```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('data/twi_data.tsv', sep='\t')
df.head()
```

```python
df.shape
```

```python
date_counts = pd.DataFrame(df.date.value_counts()).sort_index()
date_counts.head()
```

```python
plt.bar(date_counts.index, date_counts.date, width=0.7)
plt.xlabel('Date')
plt.ylabel('Number of tweets')
plt.xticks(rotation=50, ha='right')
plt.rcParams['figure.figsize'] = (10, 7)
plt.title('Number of tweets per day during month')
plt.show()          
```

```python
city_counts = pd.DataFrame(df.city.value_counts()).sort_values('city', ascending=False)
city_counts.head()
```

```python
c = 0
for n in city_counts.city:
    if n >= 100:
        c += 1
    else:
        break
        
c
```

```python
city_counts[:c]
```

```python
labels = city_counts.index[:c]
sizes = city_counts.city[:c]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```
